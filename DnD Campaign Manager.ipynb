```python
import random, itertools
```

```python
class IterCharacter(type):
    def __iter__(cls):
        return iter(cls._allCharacters)

class IterMonsters(type):
    def __iter__(cls):
        return iter(cls._allMonsters)
    
class IterNPCS(type):
    def __iter__(cls):
        return iter(cls._allNPCS)
    
class Monster(metaclass=IterMonsters):
    _allMonsters = []
    def __init__(self, Name, AC,HP, Speed, Str, Dex, Con, Int, Wis, Char, Chal, Prof):
        self._allMonsters.append(self)
        self.Name = Name
        self.AC = AC
        self.Speed = Speed
        self.Str = Str
        self.Dex = Dex
        self.Con = Con
        self.Int = Int
        self.Wis = Wis
        self.Char = Char
        self.Chal = Chal
        self.Prof = Prof
        self.Init = 0
        self.initiative = 0

    def init_order (self):
        roll = random.randint(1,20)
        self.initiative = roll + self.Init
        return self.initiative

    
class Character(metaclass=IterCharacter):
    _allCharacters = []  
    def __init__(self, Name, AC, Init, Str, Dex, Con, Int, Wis, Char, Level, Walk):
        self._allCharacters.append(self)
        self.Name = Name
        self.AC = AC
        self.Init = Init
        self.Str = Str
        self.Dex = Dex
        self.Con = Con
        self.Int = Int
        self.Wis = Wis
        self.Char = Char
        self.Level = Level
        self.Walk = Walk
        self.Items = {}
        self.Weapons = []
        self.initiative = 0
       
            
    def prof_calculator(self):
        Prof = 0
        if self.Level <= 4:
            Prof = 2
            print(Prof)
        elif self.Level <= 8 > 4:
            Prof = 3
            print(Prof)
        elif self.Level <= 12 >8:
            Prof = 4
            print(Prof)
        elif self.Level <=16 >12:
            Prof = 5
            print(Prof)
        elif self.Level <=20 >16:
            Prof = 6
            print(Prof)
        else:
            print("Level is out of range!")
    
    def add_items(self,**items):
        for key, value in items.items():
            self.Items[key] = value
    
    def remove_items(self, **items):
        for key, value in items.items():
            del self.Items[key]
        
    
    def add_weapons(self, *weapons):
        for weapon in weapons:
            self.Weapons.append(weapon)
            
    def remove_weapon(self, *weapons):
        for weapon in weapons:
            self.Weapons.remove(weapon)
    
    def init_order (self):
        roll = random.randint(1,20)
        self.initiative = roll + self.Init
        return self.initiative
        
        
            
    
class Encounter:
    def __init__(self,**Entities):
        self.Entities = Entities
        
    
    def player_sort():
        encPlayers = {}
        encMonster =  {}
        for c in Character:
            c.init_order()
            encPlayers[c.Name] = c.initiative
        for m in Monster:
            m.init_order()
            encMonster[m.Name] = m.initiative
        encPlayers.update(encMonster)
        return dict(sorted(encPlayers.items(), key=lambda item: item[1]))

    
    def dice_roll(self, *dice):
        result = 0
        for die in dice:
            print(f'You rolled a D{die}')
            if die == 20:
                roll = random.randint(1,die)
                if roll == 20:
                    print(f'{roll} Critical Success!')
                elif roll == 1:
                    print(f'{roll} Critical Fail!')
                else:
                    print(f'Your D20 result is {roll}')
            else:
                roll = random.randint(1,die)
                print(f'and your D{die}\'s result is {roll}')
                result += roll
        return (f'Your total is {result}')
```
